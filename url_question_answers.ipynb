#Load the URL

```python
from langchain.document_loaders import UnstructuredURLLoader
urls = [
    "https://en.wikipedia.org/wiki/Narendra_Modi"
]
loader2 = [UnstructuredURLLoader(urls=urls)]
```

```python
print(loader2)
```

#Splitting the Data

```python
index2 = VectorstoreIndexCreator(
    embedding=HuggingFaceEmbeddings(),
    text_splitter=CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)).from_loaders(loader2)
```

#Retrieve the answer from the question prompted by user

```python
llm2=HuggingFaceHub(repo_id="declare-lab/flan-alpaca-large", model_kwargs={"temperature":0, "max_length":512})
from langchain.chains import RetrievalQA
chain = RetrievalQA.from_chain_type(llm=llm2,
                                    chain_type="stuff",
                                    retriever=index2.vectorstore.as_retriever(),
                                    input_key="question")

# chain.run('when was narendra modi born')
```

#Give the answer based on the user question

```python
while(True):
  UserQuestion = input("Enter your Question:")
  if(UserQuestion == 'exit'):
    break
  else:
    answer = chain.run(UserQuestion)
    print("Question:", UserQuestion)
    print("Answer:", answer)
    print("--------------")
```
